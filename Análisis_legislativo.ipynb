###Inspección de nans

Los NaN se dan en las columnas de apoyo, donde los bloques no se hicieron presentes

```python
df[ df.isna().any(axis=1) ][[c for c in df.columns if (c[1]=='_' or c in['tipo_mayoria','resultado'])]]
```

Los NaN se dan en las columnas de apoyo en los casos donde los bloques mayoritarios no se presentaron a votar. Esto se da en todas (con una sola excepción) las votaciones donde se requiere un tipo de mayoria 0, lo que equivale a los tipos 'Mitad más uno' y 'Más de la mitad'.
Además, Y además en todos estos casos resulta que el acta en votación fué aprobada.

Es sabido que el hecho de no presentarse a una votación puede utilisarse como herramienta política para no dar quorum a la sesión y posponer esta y las demás votaciones que se disputen en dicha sesión. Es por esto que cualquier decisión que se tome para trabajar con estos samples introducirá un sesgo en el resultado, y reemplazar los NaNs por algún valor adecuado no es nada trivial.

Considerando que en estos casos el resultado de la votación es el mismo, Afirmativo, y que esta clase ya se encuentra sobrerepresentada, Tomo la decisión de eliminar estas filas. Además puede verse que esto no quita variabilidad a los demas features.
